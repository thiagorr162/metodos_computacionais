# NumPy

## Exercício 1
Crie um vetor com 20 números inteiros aleatórios entre 0 e 100.  
- Mostre o vetor.  
- Calcule média, desvio-padrão e valor máximo.  

## Exercício 2
Gere uma matriz $5 \times 5$ com valores da distribuição normal padrão.  
- Calcule a soma de cada linha e de cada coluna.  
- Extraia a diagonal principal.  

## Exercício 3
Implemente uma função que recebe dois vetores NumPy e retorna:  
- Produto escalar (`np.dot`).  
- Distância euclidiana entre eles (`np.linalg.norm`).  


## Exercício 4
Crie uma matriz quadrada $A \in \mathbb{R}^{4 \times 4}$ com valores inteiros aleatórios entre $-5$ e $5$.  
- Verifique se $A$ é invertível calculando $\det(A)$ (`np.linalg.det`).  
- Caso seja, calcule $A^{-1}$ (`np.linalg.inv`) e verifique se $A \cdot A^{-1} = I$.  

## Exercício 5
Gere uma matriz simétrica $B \in \mathbb{R}^{5 \times 5}$ a partir de uma matriz aleatória $M$ definida por $B = \tfrac{1}{2}(M + M^T)$.  
- Calcule os autovalores e autovetores (`np.linalg.eig`).  
- Verifique numericamente a relação $Bv = \lambda v$.  

## Exercício 6
Considere um sistema linear $Ax = y$, com $A \in \mathbb{R}^{3 \times 3}$ e $y \in \mathbb{R}^3$.  
- Gere $A$ de forma aleatória e $y$ como um vetor de inteiros.  
- Resolva o sistema (`np.linalg.solve`).  
- Verifique a solução computando $\|Ax - y\|_2$ (`np.linalg.norm`).  

## Exercício 7
Simule 100 vetores bidimensionais com distribuição normal padrão e forme a matriz de covariância empírica $S$.  
- Calcule a matriz de covariância (`np.cov`).  
- Encontre os autovalores e autovetores de $S$ (`np.linalg.eig`).  
- Mostre que os autovetores correspondem às direções principais da variabilidade dos dados.
