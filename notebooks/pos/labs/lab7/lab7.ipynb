# Exerccios  Cadeias de Markov

Considere as duas cadeias de Markov representadas na Figura abaixo.



1. Escreva a matriz de transio $Q$ de cada cadeia.  

2. Usando `numpy`, calcule as potncias $Q^n$ para diferentes valores de $n$.  

3. Implemente uma simulao de trajetrias dessas cadeias de Markov.  
   - Use as tcnicas de **paralelizao** vistas em sala para simular muitas cadeias de forma eficiente (por exemplo, milhares de trajetrias de comprimento $n$).  
   - A partir das simulaes, estime a **probabilidade emprica** de estar em cada estado aps $n$ passos.  

4. Compare os resultados empricos obtidos nas simulaes com as probabilidades tericas calculadas no item (2).  

5. Para as duas cadeias da Figura, fixe diferentes condies iniciais:  
   - comeando em cada estado individualmente ($X_0 = i$);  
   - comeando de uma distribuio inicial uniforme.  

6. Usando `numpy`, calcule a distribuio de probabilidades de estar em cada estado aps **500 passos**.  

7. Compare os resultados obtidos para diferentes condies iniciais.
