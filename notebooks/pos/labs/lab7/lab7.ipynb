# Exercícios – Cadeias de Markov

Considere as duas cadeias de Markov representadas na Figura abaixo.



1. Escreva a matriz de transição $Q$ de cada cadeia.  

2. Usando `numpy`, calcule as potências $Q^n$ para diferentes valores de $n$.  

3. Implemente uma simulação de trajetórias dessas cadeias de Markov.  
   - Use as técnicas de **paralelização** vistas em sala para simular muitas cadeias de forma eficiente (por exemplo, milhares de trajetórias de comprimento $n$).  
   - A partir das simulações, estime a **probabilidade empírica** de estar em cada estado após $n$ passos.  

4. Compare os resultados empíricos obtidos nas simulações com as probabilidades teóricas calculadas no item (2).  

5. Para as duas cadeias da Figura, fixe diferentes condições iniciais:  
   - começando em cada estado individualmente ($X_0 = i$);  
   - começando de uma distribuição inicial uniforme.  

6. Usando `numpy`, calcule a distribuição de probabilidades de estar em cada estado após **500 passos**.  

7. Compare os resultados obtidos para diferentes condições iniciais.
